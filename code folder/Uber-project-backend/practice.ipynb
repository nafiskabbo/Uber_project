```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371 # km
    return c * r

df['trip_distance'] = df.apply(lambda x: haversine(x['pickup_longitude'], x['pickup_latitude'],
                                                     x['dropoff_longitude'], x['dropoff_latitude']), axis=1)
```

```python

df = df[(df["trip_distance"] > 0) & (df["trip_distance"] <= 50)]
```

```python
def manhattan_distance(lat1, lon1, lat2, lon2):
    a = haversine(lat1, lon1, lat1, lon2)
    b = haversine(lat1, lon1, lat2, lon1)
    return a + b

df['manhattan_distance'] = df.apply(lambda x: manhattan_distance(x['pickup_latitude'], x['pickup_longitude'], 
                                                                 x['dropoff_latitude'], x['dropoff_longitude']), axis=1)
```
